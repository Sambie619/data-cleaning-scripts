```python
#dropping unwanted columns
data.drop(columns=['Unnamed: 0','TestPrep', 'ParentMaritalStatus', 'PracticeSport', 'IsFirstChild'],inplace=True)
```

```python
data
```

```python
#Renaming Columns 
data.rename(columns={"ParentEduc":"ParentEducation","NrSiblings":"No: of Siblings","WklyStudyHours":"WeeklyStudyHours"},inplace=True)
data
```

```python
#Finding duplicated rows
data.duplicated().sum()
```

```python
#Removing the 9 duplicate rows
data.drop_duplicates(inplace=True)
data
```

```python
#Checking for NaN values count
data.isna().sum().sum()
```

```python
#Dropping those NaN values crucial for accurate data analysis
data.dropna(inplace=True)
data.isna().sum()
```

```python
#changing some Categorical values to numerical values,crucial step in data preprocessing for ML
data["Gender"]=data["Gender"].apply(lambda x:1 if x=="female"else 0)
data
```

```python
#index is messed up,let's fix that
data.reset_index(inplace=True)
data
```

```python
#fixing the mismatched data types of Columns
type(data["No: of Siblings"][0])
```

```python
data["No: of Siblings"]=data["No: of Siblings"].astype(int)
type(data["No: of Siblings"][0])
```

```python
#Exporting cleaned data to csv,excel files
data.to_csv("Cleaned_csv_data.csv")
data.to_excel("Cleaned_excel_data.xlsx")
```
