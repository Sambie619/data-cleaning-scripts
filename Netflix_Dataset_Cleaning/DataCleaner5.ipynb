```python
#let's drop unnecessary columns 
data.drop(columns='rating',inplace=True)
data
```

```python
#Flixing structural errors like wrong column name 
col=[]
for i in data.columns:
    col.append(i.capitalize())
col
```

```python
data.columns=col
```

```python
data
```

```python
data.duplicated().sum()
```

```python
#for removing duplicates use 
data.drop_duplicates(inplace=True)
```

```python
#Handling missing data 
data.isna().sum()
```

```python
#if there was NaN cols 
data.dropna(inplace=True)
```

```python
#correcting inaccurate values/typo's 
data['Show_id'].unique()
#data['Show_id'].nunique()
```

```python
#change s1,s2,s3...to 1,2,3... 
data['Show_id']=data['Show_id'].apply(lambda x:x.split("s")[1])
```

```python
data['Date_added']=data['Date_added'].apply(lambda x:x.replace("/","-"))
data['Date_added']
```
