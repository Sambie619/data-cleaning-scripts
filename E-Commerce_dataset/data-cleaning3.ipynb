```python
#dropping illogical columns
data.drop(columns="Unnamed: 0",inplace=True)
data
```

```python
#renaming columns for better understaning
data.rename(columns={"Deatils":"Details"},inplace=True)
data.columns
```

```python
#Figuring out total duplicate columns
#data.duplicated().sum()
data.duplicated().value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
#visualizing duplicate rows to toal rows
data.duplicated().value_counts().plot(kind="bar")
plt.title("Duplicate Rows")
plt.xlabel("Duplicate or Not")
plt.ylabel("Total Rows")
plt.show()
```

```python
#Dropping duplicate rows
data.drop_duplicates(inplace=True)
data.duplicated().sum()
```

```python
data=data.reset_index().drop(columns="index")
```

```python
data
```

```python
#dropping NaN values,Data Cleaning
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
#Capitalizing Brandname columns
data["BrandName"]=data["BrandName"].apply(lambda x:x.capitalize()) 
```

```python
data
```

```python
data["Details"]=data["Details"].apply(lambda x:x.capitalize()) 
data
```

```python
#I've seen that some of the columns in the Sizes col are entered as Nan,its not actually a NaN(undefined)
#but entered as it,just look at its type
type(data["Sizes"][27926])
```

```python
data["Sizes"]=data["Sizes"].replace("Nan","NA Values")
```

```python
data
```

```python
#let's filter data by removing rows with NA Valued
data=data[data["Sizes"]!="NA Values"]
```

```python
data
```

```python
data["Discount"]=data["Discount"].replace("Nan","No Discount")
```

```python
data
```

```python
#from above dataset view we can see that total rows not matchign with the index,so reset index
data=data.reset_index().drop(columns="index")
```

```python
#exporting data to excel,csv
data.to_csv("cleaned_csv.csv")
data.to_excel("cleaned_excel.xlsx")
```
